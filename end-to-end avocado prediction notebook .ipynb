### To pick the right model we'll use two functions:
* one to estimate the `r2_score`, `mean_absolute_error`and `mean_squared_error` with the split from above i.e. the X_train,X_test,y_train, y_test 
* the other will use the `cross_validate` on the same three metrics above with the whole dataset ensuring we didn't get a lucky split

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge 
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import cross_val_score 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_validate
```

```python
def model_comparison(X_train, y_train,X_test,y_test,models):
    """
    Compares the different models based on  r^2,mean_squared_error, and mean_abolute error and 
     score between given models dict  and retruns a dictionry with the results 
    """
    model_metrics={}
    for model_name, model in models.items():
        print("Creating a model using {}".format(model_name))
        model.fit(X_train,y_train)
        print("Done")
        measures =[]
        measures.append( r2_score(y_test,model.predict(X_test)))
        measures.append(mean_absolute_error(y_test,model.predict(X_test)))
        measures.append(mean_squared_error(y_test,model.predict(X_test)))
        model_metrics[model_name] = measures
    return model_metrics
```

```python
def model_cross_val(X,y,models): 
    """
    Estmates the models using cross val score an rturns a dict containg the r2_score, MAE, MSE
    """
    metrics = {}
    for model_name, model in models.items():
        print("Training model {}".format(model_name))
        cross_val_result = cross_validate(model,
                                                                    X,
                                                                      y,
                                                                    cv=5,
                                                                 scoring= ("r2","neg_mean_squared_error","neg_mean_absolute_error"))
        print("Done")
        values = [] # An empty list which stores the values of the metrics and then it's aded to the metric dict
        values.append(cross_val_result["test_r2"].mean())
        values.append(cross_val_result["test_neg_mean_absolute_error"].mean())
        values.append(cross_val_result["test_neg_mean_squared_error"].mean())
        metrics[model_name] = values
        
    return metrics 
        
```

```python
models = {"RandomForestRegressor":RandomForestRegressor(),
                    "SGDRegressor":SGDRegressor(),
                   "Lasso":Lasso(),
                    "Ridge":Ridge(),
                    "XGboost":XGBRegressor(),
                     "Catboost":CatBoostRegressor()}
```

```python
%time 
metrics = model_comparison(X_train, y_train, X_test, y_test,models)
metrics
```

### Now it's time for hyperparameter tuning 

> Since we're using `catboost` we'll input the data without converting the fetures into catgorical 

```python
avocado_df = pd.read_csv("edited.csv")

avocado_df.head()
```

```python
X_cat = avocado_df.drop("AveragePrice",axis=1)
y_cat = avocado_df.AveragePrice
X_cat.head()
```

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
X_cat_train, X_cat_test, y_cat_train, y_cat_test = train_test_split(X_cat, y_cat,test_size=0.2, random_state=55)

model_1 = CatBoostRegressor(cat_features=["type","region"],verbose=1)

model_1.fit(X_cat_train,y_cat_train)

r2_score(y_cat_test,model_1.predict(X_cat_test))
```

```python
r2_score(y_cat_train,model_1.predict(X_cat_train))
```

```python
from sklearn.model_selection import RandomizedSearchCV 

rand_model = CatBoostRegressor(cat_features=["type","region"],
                                                                            verbose=0,
                                                                          random_seed=333)

rand_params = {"learning_rate":np.linspace(0.35,0.42),
                           "depth" : np.arange(4,10,1),
                           "l2_leaf_reg": np.arange(0.1,10,100),
                          "iterations": np.arange(900,1100,50)}

rand_search_model = RandomizedSearchCV(rand_model, 
                                      rand_params, 
                                      n_iter= 20,
                                      cv=3,
                                      verbose=1.0)

rand_search_model.fit(X_cat_train,y_cat_train)
```

```python
r2_score(y_cat_test,rand_search_model.predict(X_cat_test))
```

```python
r2_score(y_cat_train,rand_search_model.predict(X_cat_train))
```

```python
# A line for fixing autocomplete
%config Completer.use_jedi = False
```

```python
def error_metrics (y_true, y_preds):
    """
    Takes X&y as inputs and rteurns r^2 scor, MAE, and MSE 
    """
    metrics = {"r2_score":0,"MAE":0,"MSE":0}
    metrics["r2_score"] = r2_score(y_true,y_preds)
    metrics["MAE"] =mean_absolute_error(y_true,y_preds)
    metrics["MSE"] = mean_squared_error(y_true,y_preds)
    
    return metrics 
```

```python
y_test_preds = rand_search_model.predict(X_cat_test)
test_metrics = error_metrics(y_cat_test,y_test_preds)
y_train_preds = rand_search_model.predict(X_cat_train)
train_metrics = error_metrics(y_cat_train,y_train_preds)

plt.style.use("seaborn-whitegrid")
fig, (ax1,ax2) = plt.subplots(nrows=1,
                                                     ncols=2,
                             figsize=(10,5))

ax1.bar(train_metrics.keys(),train_metrics.values())
ax2.bar(test_metrics.keys(),train_metrics.values())
fig.suptitle("a Comparison between the test metrics and the train metrics")
ax1.set(title=("Train"))
ax2.set(title=("Test"));
```

```python
for k in train_metrics.keys():
    print("Test {} is:  {:.4f} ".format(k,test_metrics[k]))
    print("Train {} is: {:.4f}".format(k,train_metrics[k]))
```

It looks like the model is overfitting let's try setting the `early_stopping_round` parameter to be equal to 10 

```python
# Getting the best hyperparameters to fit in the next model 
rand_search_model.best_params_
```

```python
model2 = CatBoostRegressor(early_stopping_rounds=10,
                                                    learning_rate=0.35,
                                                      l2_leaf_reg = 0.1,
                                                          iterations=1000,
                                                          depth = 6 ,
                                                          cat_features=["type","region"])
model2.fit(X_cat_train,
                   y_cat_train,
                   plot=True, # In order to see the difference between the test and train sets 
                   eval_set=(X_cat_test, y_cat_test))
```

```python
# A quick comparison to see the performance of the model 
y_preds_test = model2.predict(X_cat_test)
y_preds_train = model2.predict(X_cat_train)
test_metrics = error_metrics(y_cat_test, y_preds_test)
train_metrics = error_metrics(y_cat_train, y_preds_train)

for k in test_metrics.keys():
    print("Test's {} is : {:.2f} \t as opoesed to Train's {} which is: {:.2f}".format(k,test_metrics[k],k,train_metrics[k]))
```
